```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,roc_curve,accuracy_score, classification_report
```

```python
#Checking Missing Values
def fill_nan(df):
    for index, column in enumerate(df.columns[:9]):
        # print(index, column)
        df[column] = df[column].fillna(df.groupby('Potability')[column].transform('mean'))
    return df
        
df = fill_nan(df_raw)
df.isna().sum()  
df.describe()
```

```python
scaler = MinMaxScaler()
array = scaler.fit_transform(df.to_numpy())
df2 = pd.DataFrame(array, columns = ['ph','Hardness','Solids','Cloramines', 'Sulphate', 'Conductivity', 'Organic_carbon', 'Trihalomethanes','Turbidity', 'Potability'])
df2.describe()
```

```python
def PCA_manual(X , num_components):
     
    #Step-1
    X_meaned = X - np.mean(X , axis = 0)
     
    #Step-2
    cov_mat = np.cov(X_meaned , rowvar = False)
     
    #Step-3
    eigen_values , eigen_vectors = np.linalg.eigh(cov_mat)
     
    #Step-4
    sorted_index = np.argsort(eigen_values)[::-1]
    sorted_eigenvalue = eigen_values[sorted_index]
    sorted_eigenvectors = eigen_vectors[:,sorted_index]
     
    #Step-5
    eigenvector_subset = sorted_eigenvectors[:,0:num_components]
     
    #Step-6
    X_reduced = np.dot(eigenvector_subset.transpose() , X_meaned.transpose() ).transpose()
     
    return X_reduced
```

```python
#prepare the data
X = df2.drop(['Potability'], axis = 1)
y = df2['Potability']
 
#Applying it to PCA function
mat_reduced = PCA_manual(X , 5)
 
#Creating a Pandas DataFrame of reduced Dataset
df_scratch = pd.DataFrame(mat_reduced , columns = ['PC1','PC2','PC3','PC4','PC5'])
 
#Concat it with target variable to create a complete Dataset
df_scratch = pd.concat([df_scratch , pd.DataFrame(y)] , axis = 1)

df_scratch.head(5)
```

```python
pca = PCA(n_components=5)
pca.fit(X)
mat = pca.transform(X)
df_built_in = pd.DataFrame(mat, columns = ['PC1','PC2','PC3','PC4','PC5'])
df_built_in = pd.concat([df_built_in , pd.DataFrame(y)] , axis = 1)
print(df_built_in)

```

```python
#Dividing dataset into traing and testing set with 30-70 ratio 
#Dropping last column (output)
X_scratch = df_scratch.drop(['Potability'], axis = 1)
y_scratch = df_scratch['Potability']

X_built_in = df_built_in.drop(['Potability'], axis = 1)
y_built_in = df_built_in['Potability']

# Splitting
X_train_scratch, X_test_scratch, y_train_scratch, y_test_scratch = train_test_split(X_scratch,y_scratch, test_size=0.3, random_state=1111, stratify=y_scratch) #stratify=y
X_train_built_in, X_test_built_in, y_train_built_in, y_test_built_in = train_test_split(X_built_in,y_built_in, test_size=0.3, random_state=1111, stratify=y_built_in) #stratify=y
```

```python
#KNN 
from sklearn.neighbors import KNeighborsClassifier 

k = range(1,20,1)
testing_accuracy_scratch = []
training_accuracy_scratch = []

testing_accuracy_built_in = []
training_accuracy_built_in = []
score = 0

#scratch
for i in k:
    knn= KNeighborsClassifier(n_neighbors=i, metric='minkowski', p=3 )  
    knn.fit(X_train_scratch, y_train_scratch)  
    y_pred_train = knn.predict(X_train_scratch)
    training_accuracy_scratch.append(accuracy_score(y_train_scratch, y_pred_train))
    
    y_pred_test = knn.predict(X_test_scratch)
    acc_score = accuracy_score(y_test_scratch,y_pred_test)
    testing_accuracy_scratch.append(acc_score)
    
    if score < acc_score:
        score = acc_score
        best_k = i
        
print('Best Accuracy Score for PCA from scratch', score, 'Best K-Score', best_k)

#Built_in
for i in k:
    knn= KNeighborsClassifier(n_neighbors=i, metric='minkowski', p=3 )  
    knn.fit(X_train_built_in, y_train_built_in)  
    y_pred_train = knn.predict(X_train_built_in)
    training_accuracy_built_in.append(accuracy_score(y_train_built_in, y_pred_train))
    
    y_pred_test = knn.predict(X_test_built_in)
    acc_score = accuracy_score(y_test_built_in,y_pred_test)
    testing_accuracy_built_in.append(acc_score)
    if score < acc_score:
        score = acc_score
        best_k = i

print('Best Accuracy Score for PCA from built -in', score, 'Best K-Score', best_k)
```
